### 特徵解釋
- InvoiceNo: Invoice number. Nominal, a 6-digit integral number uniquely assigned to each transaction. If this code starts with letter 'c', it indicates a cancellation.
- StockCode: Product (item) code. Nominal, a 5-digit integral number uniquely assigned to each distinct product.
- Description: Product (item) name. Nominal.
- Quantity: The quantities of each product (item) per transaction. Numeric.
- InvoiceDate: Invice Date and time. Numeric, the day and time when each transaction was generated.
- UnitPrice: Unit price. Numeric, Product price per unit in sterling.
- CustomerID: Customer number. Nominal, a 5-digit integral number uniquely assigned to each customer.
- Country: Country name. Nominal, the name of the country where each customer resides.

```python
print('原先包含退貨數量', len(df))
df = df.loc[df['Quantity'] > 0]

print('去除退貨總數量:', len(df))
```

### 去除NaN資料

```python
df['CustomerID'].isna().sum()     # 沒有就無法分析, 故remove
```

```python
df = df[~df['CustomerID'].isna()]
# or df = df.dropna(subset=['CustomerID'])
df.shape
```

### 建立消費者矩陣(共線矩陣)

```python
customer_item_matrix = df.pivot_table(
    index='CustomerID',
    columns='StockCode',
    values='Quantity',
    aggfunc='sum'
)
customer_item_matrix
```

```python
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
customer_item_matrix
```

### 協同過濾 - user-based
1. 選定一個`消費者i`
2. 計算`消費者i`與`其他rows(消費者)之間的相似性`，找出`最相似的n個消費者`
3. 並計算消費者i沒有消費過的商品中，最相似的n個消費者對於該商品的加權平均
    - 如果是連續值就是分數與相似性程度標準化的加權
    - 如果是離散值就是點擊與否與相似性標準化的加權
    - 附註: 其實也可以一視同仁，將n個視為相同權重。

```python
# user-based
from sklearn.metrics.pairwise import cosine_similarity

user_user_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix)
)
user_user_sim_matrix
# cosine_similarity(customer_item_matrix)
```

```python
# index與column很不直覺，改一下

user_user_sim_matrix.columns = customer_item_matrix.index
user_user_sim_matrix
```

```python
user_user_sim_matrix['CustomerID'] = customer_item_matrix.index        # 新增column
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')    # 將這個column設為index
```

```python
user_user_sim_matrix
```

```python
# 12346.0 這個消費者與其他人的相似度
user_user_sim_matrix.loc[12346.0].sort_values(ascending=False)
```

```python
# 12350.0 買的商品
items_bought_by_A = set(customer_item_matrix.loc[12350.0].iloc[
    customer_item_matrix.loc[12350.0].to_numpy().nonzero()
].index)
items_bought_by_A
```

```python
# 17935.0 買的商品
items_bought_by_B = set(customer_item_matrix.loc[17935.0].iloc[
    customer_item_matrix.loc[17935.0].to_numpy().nonzero() # np.ndarray.nonzero() ---> return indices of nonzero elements
].index)
items_bought_by_B
```

```python
items_to_recommend_to_B = items_bought_by_A - items_bought_by_B
items_to_recommend_to_B
```

```python
# 顯示推薦給B的商品以及其敘述
df.loc[
    df['StockCode'].isin(items_to_recommend_to_B), 
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode')    # 需要drop_duplicates是因為df有許多同樣的商品購買
```

### 協同過濾 - item-based

```python
# 商品相似矩陣
item_item_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix.T)
)

item_item_sim_matrix
```

```python
# 將columns改名
item_item_sim_matrix.columns = customer_item_matrix.T.index
item_item_sim_matrix
```

```python
# 將index也改名
item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
item_item_sim_matrix
```

```python
# 23166的相似10個產品
top_10_similar_items = list(
    item_item_sim_matrix.loc[23166].sort_values(ascending=False).iloc[:10].index
)

top_10_similar_items
```

```python
# 顯示出那10個產品以及其敘述

df.loc[
    df['StockCode'].isin(top_10_similar_items),
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode').loc[top_10_similar_items]
```

## 重新整理並擴展到 top-n 加權版本

```python
def get_cocurrence_matrix(df, index_name, column_name, values_name):
    """
        從原始點擊、購買紀錄轉換成共線矩陣，
        可以代表 `消費者` 與 `產品` 的點擊/購買紀錄，
        方便後續透過紀錄計算相似性。
    """
    customer_item_matrix = df.pivot_table(
        index=index_name,
        columns=column_name,
        values=values_name,
        aggfunc='sum'
    )
    customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
    return customer_item_matrix


def get_similarity_matrix(customer_item_matrix, sim_type='user', user_column='CustomerID', item_column='StockCode'):
    """
        取得user-based or item-based 相似矩陣，用於後續找出top-n使用者、商品。
    """
    if sim_type == 'user':
        user_user_sim_matrix = pd.DataFrame(
            cosine_similarity(customer_item_matrix)
        )
        user_user_sim_matrix.columns = customer_item_matrix.index
        user_user_sim_matrix[user_column] = customer_item_matrix.index        # 新增column
        user_user_sim_matrix = user_user_sim_matrix.set_index(user_column)    # 將這個column設為index
        return user_user_sim_matrix
    else:
        item_item_sim_matrix = pd.DataFrame(
            cosine_similarity(customer_item_matrix.T)
        )
        item_item_sim_matrix.columns = customer_item_matrix.T.index
        item_item_sim_matrix[item_column] = customer_item_matrix.T.index
        item_item_sim_matrix = item_item_sim_matrix.set_index(item_column)
        return item_item_sim_matrix
    

def get_top_n_simi_from_matrix(simi_matrix, top_n, id_):
    """
        從相似矩陣找到 Top-n 相似消費者/產品
    """
    top_n_similar_items = dict(
        simi_matrix.loc[id_].sort_values(ascending=False).iloc[1:top_n+1]#.index
    )
    return top_n_similar_items


def get_recommend_from_top_n(customer_item_matrix, id_, top_n_similar_items, top_k):
    """
        根據 user_i 的未購買產品，
        透過該 top-n 相似產品/使用者的購買紀錄加權得出購買機率或者評分，
        再根據該評分排序選出 top_k 當作最終推薦。
    """
    items_not_bought_by_A = list((customer_item_matrix.loc[id_] != 0).index)
    top_n_user_to_unbought = customer_item_matrix.loc[list(top_n_similar_items.keys())][items_not_bought_by_A]
    item2value = {}
    for e in top_n_user_to_unbought:
        # e = column 
        # top_n_user_to_unbought[e] = 相似使用者購買紀錄
        value = (top_n_user_to_unbought[e] * np.array(list(top_n_similar_items.values()))).sum()
        item2value[e] = value
    return sorted(item2value.items(), key=lambda x :x[1], reverse=True)[:top_k]


def get_recommended_items_descripiton(top_k_product2value, id_):
    """
        取得推薦商品id以及描述
    """
    rec_products = [e[0] for e in top_k_product2value]

    print(f'消費者: {id_} 受到推薦的商品如下: ')
    for product_id in rec_products:

        print(f"商品id: {product_id}, 商品描述: {df[df['StockCode'] == product_id].iloc[0]['Description']}")
    return rec_products

def get_recommend_by_user_id():
    """
        user-based 已完成；item-basd 尚未。
    """
    pass
```

```python
## 超參數
index_name='CustomerID'
column_name='StockCode'
values_name='Quantity'
sim_type = 'user'

## 1. 找出共現矩陣、相似矩陣
customer_item_matrix = get_cocurrence_matrix(df, index_name, column_name, values_name)
user_sim_matrix = get_similarity_matrix(customer_item_matrix, sim_type)
user_sim_matrix
```

```python
# 2. 找出 top-n 相似的使用者

top_n = 10
id_ = 12346.0

top_n_similar_items = get_top_n_simi_from_matrix(user_sim_matrix, top_n, id_)
top_n_similar_items
```

```python
## 3. 透過 top-n 相似使用者推薦 top-k 相似物品
top_k = 10

top_k_product2value = get_recommend_from_top_n(customer_item_matrix, id_, top_n_similar_items, top_k)
top_k_product2value
```

```python
## 查看一下推薦的商品都是哪一些

get_recommended_items_descripiton(top_k_product2value, id_)
```

> 可以發現都是家居用品，如儲存罐(大多)、籃子、凳子、毛巾、烘焙核，推薦出來的商品相似性是很高的！
